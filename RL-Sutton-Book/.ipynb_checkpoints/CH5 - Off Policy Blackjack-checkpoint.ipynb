```python
episode = episode_set[19]
r = -1

# IMPORTANCE SAMPLING RATIO

# probability of taking action a at state s
def p_pi(a,s):
    return (s>19)^(a)
    
def p_b(a,s):
    return 0.5

def ISR(episode):
    isr = 1
    for (s,a) in episode:
        isr *= p_pi(a,s)/p_b(a,s)
    return isr
```
